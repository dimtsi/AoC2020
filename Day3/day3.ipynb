```python
with open('input.txt', "r") as f:
    matrix = [line.rstrip() for line in f]
matrix[:5]
```

```python
#repeat

matrix_rep = list(map(lambda x: x*100, matrix))
matrix_rep[0]
```

```python
tree_count=0
i=0

for j, line in enumerate(matrix_rep):
    if line[i] == "#":
        tree_count += 1
    i = i+3
    
print(tree_count)
     
```

```python
# part2
from functools import reduce
def count_trees(right_step, down_step):
    tree_count = 0
    i = 0
    j = 0
    
    while(i < len(matrix_rep)):
        if matrix_rep[i][j] == "#":
            tree_count += 1
        i += down_step
        j += right_step
    return tree_count


step_args = [(1, 1), (3, 1), (5, 1), (7, 1), (1, 2)]

counts = [count_trees(*arg) for arg in step_args]
totals = reduce(lambda x,y: x*y, counts)
totals
```
